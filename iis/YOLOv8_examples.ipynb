```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.pt')  # load a pretrained model (recommended for training)
results = model('https://ultralytics.com/images/bus.jpg')
for result in results:
    boxes = result.boxes  # Boxes object for bounding box outputs
    masks = result.masks  # Masks object for segmentation masks outputs
    keypoints = result.keypoints  # Keypoints object for pose outputs
    probs = result.probs  # Probs object for classification outputs
    result.show()  # display to screen
    result.save(filename='result.jpg')

for result in results:
    boxes = result.boxes  # объект Boxes для вывода ограничивающих рамок
    for box in boxes.xyxy:
        # box.xyxy содержит координаты левого верхнего угла (x1, y1) и правого нижнего угла (x2, y2) каждой рамки
        x1, y1, x2, y2 = box
        print("Координаты рамки: x1 =", x1, "y1 =", y1, "x2 =", x2, "y2 =", y2)
```
